```python
def get_graph(**kwargs):
    nvid_list = ndf['nvid'].astype(str).unique()
    nvid = kwargs.get('nvid', 'all')  # Default to 'all' if no nvid is passed
    
    if nvid == 'all':
        # Group by 'nvid' and sum the 'Battery_Capacity_AH'
        grouped_data = ndf.groupby('nvid')['Battery_Capacity_AH'].sum()
    elif nvid in nvid_list:
        # Filter for the specific nvid and group
        df = ndf[ndf['nvid'] == nvid]
        grouped_data = df.groupby('nvid')['Battery_Capacity_AH'].sum()
    else:
        print(f"Invalid nvid: {nvid}. Please choose from {list(nvid_list)} or use 'all'.")
        return
    
    # Drop NaN and zero values
    grouped_data = grouped_data.dropna()
    grouped_data = grouped_data[grouped_data != 0]

    if grouped_data.empty:
        print(f"No valid data to plot for nvid: {nvid}")
    else:
        # Plot the bar graph for non-null and non-zero values
        grouped_data.plot(kind='bar', figsize=(10, 6), color='skyblue')
        plt.title(f"Battery Capacity {'for All nvids' if nvid == 'all' else f'for nvid: {nvid}'}")
        plt.xlabel('nvid')
        plt.ylabel('Battery Capacity (AH)')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    
    # Print grouped data
    print(f"Grouped data {'for all nvids' if nvid == 'all' else f'for nvid: {nvid}'}:")
    print(grouped_data)
    print('#' * 15, '\n')
```
